# Data Visualization using Pandas

```python
import pandas as pd
import numpy as np
```

## Line Plot
Line Plot are useful for showing trends over intervals.

```python
# Sample data: Daily temperature for a city over a week
data = {
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'temperature': [25, 26, 27, 28, 29, 30, 31]  # Temperature in Celsius
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
# Plot the line chart using Pandas plotting API
df.plot()
```

Line Plot is the default plot type in Pandas.

```python
# Added few more properties x,y axis, marker, title and color
df.plot(
    x='day', 
    y='temperature',
    kind='line', 
    marker='o', 
    linestyle='-', 
    color='b', 
    title='Daily Temperature Variation'
)
```

Line Plot shows the daily temperature variation for the given week without needing of external modules/api's.

## Bar Plot
Bar plots are useful for comparing quantities corresponding to different groups.

```python
# Sample data: Monthly sales data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [10000, 12000, 9000, 11000, 10500]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
df
```

```python
# Plot the bar chart using Pandas plotting API
df.plot(x='Month', y='Sales', kind='bar')
```

```python
df.plot(x='Month', y='Sales', kind='bar', color='skyblue', title='Monthly Sales')
```

```python
# Plot the horizontal bar chart 
df.plot(x='Month', y='Sales', kind='barh', color='lightgreen', title='Monthly Sales')
```

## Histogram 

```python
# Create a DataFrame from the data
df = pd.DataFrame(
    {
        'Age': [25, 30, 35, 40, 45, 35, 35, 40, 35, 30, 25, 40, 25, 30]
    }
)
```

```python
# Plot the histogram 
df.plot(kind='hist', bins=10, color='skyblue', title='Age Distribution', density=True)
```

```python
df['Age'].value_counts()
```

```python
df['Age'].value_counts().sort_index().plot(kind='bar', color='skyblue', title='Age Distribution')
```

## Box Plot

```python
# Sample data: Monthly sales data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [10000, 12000, 9000, 11000, 10500]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
df
```

```python
# Plot the box plot
df.plot(kind='box', y='Sales', title='Monthly Sales')
```

```python
# Plot the box plot with additional attributes
boxprops = dict(linestyle='--', linewidth=2, color='blue')  # Properties for the box
whiskerprops = dict(linestyle='-.', linewidth=2, color='green')  # Properties for the whiskers
medianprops = dict(linestyle='-', linewidth=2, color='red')  # Properties for the median line
capprops = dict(linestyle=':', linewidth=2, color='purple')  # Properties for the caps

df.plot(kind='box', y='Sales', color='lightgreen', title='Monthly Sales',
        boxprops=boxprops, whiskerprops=whiskerprops,
        medianprops=medianprops, capprops=capprops)
```

```python
# Plot the horizontal box plot
df.plot(kind='box', y='Sales', color='lightgreen', title='Monthly Sales',
        boxprops=boxprops, whiskerprops=whiskerprops,
        medianprops=medianprops, capprops=capprops, vert=False)
```

```python
df.plot(kind='box', y='Sales', color='lightgreen', title='Monthly Sales',
        showmeans=True, meanprops={'marker': 'o', 'markerfacecolor': 'blue', 'markeredgecolor': 'black'},
        showfliers=True, showcaps=True, patch_artist=True, widths=0.5)
```

## Area Plot

```python
df.plot(x='Month', y='Sales', kind='area', title='Monthly Sales (Area Plot)')
```

```python
# Plot the area plot 
df.plot(x='Month', y='Sales', kind='area', color='skyblue', title='Monthly Sales (Area Plot)')
```

```python
df.plot(x='Month', y='Sales', kind='area', color='skyblue', alpha=0.5, stacked=False,
        title='Monthly Sales (Area Plot)', linewidth=2, linestyle='--')
```

```python
df.plot(x='Month', y='Sales', kind='area', colormap='viridis', title='Monthly Sales (Area Plot)')
```

## Kernel Density Estimate plot

```python
# Create a DataFrame from the data
df = pd.DataFrame(
    {
        'Age': [25, 30, 35, 40, 45, 35, 35, 40, 35, 30, 25, 40, 25, 30]
    }
)
```

```python
# Plot the KDE plot 
df['Age'].plot(kind='kde', color='orange', title='Age Distribution (KDE)')
```

```python
df['Age'].plot(kind='kde', color='orange', title='Age Distribution (KDE)',
               linestyle='--', linewidth=2, alpha=0.7, xlim=(10, 70))
```

## Scatter plot

```python
# Sample data: Height and Weight
data = {
    'Height': [160, 165, 170, 175, 180],
    'Weight': [60, 65, 70, 75, 80]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
df.plot(x='Height', y='Weight', kind='scatter')
```

```python
df.plot(x='Height', y='Weight', kind='scatter', color='green', title='Height vs Weight')
```

```python
df.plot(x='Height', y='Weight', kind='scatter', color='green', title='Height vs Weight',
        marker='o', s=100, alpha=0.7)
```

## Hexbin plot

```python
# Plot the hexbin plot 
df.plot(x='Height', y='Weight', kind='hexbin', gridsize=20, cmap='Blues', title='Height vs Weight (Hexbin Plot)')
```
