# Seaborn
- Seaborn is a Python data visualization library based on Matplotlib.
- It provides a high-level interface for creating attractive and informative statistical graphics. 
- Seaborn is built on top of Matplotlib and tightly integrated with Pandas data structures, making it particularly convenient for visualizing data stored in DataFrames.
- It is easier to create informative and visually appealing plots for data analysis and exploration.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn comes with several built-in datasets that are useful for practicing data visualization and exploring various plotting techniques. 

```python
str(sns.get_dataset_names())
```

## Tips Dataset

```python
tips_df = sns.load_dataset('tips')
tips_df
```

## Create a histogram of total bill distribution

```python
sns.histplot(tips_df['total_bill'])
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(tips_df['total_bill'], bins=20, color='skyblue')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Set the style of the plot
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(tips_df['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

##  Create a scatter plot of total bill vs. tip amount

```python
sns.scatterplot(x = tips_df.total_bill, y=tips_df.tip, data = tips_df, hue= 'smoker')
```

A scatter plot showing the relationship between total bill and tip amount, colored by the smoker of the payer.

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex', palette='Set1', alpha=0.7)
plt.title('Total Bill vs. Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.legend(title='Sex')
plt.show()
```

Here plot a scatter plot showing the relationship between total bill and tip amount, colored by the sex of the payer.

```python
sns.relplot(x = 'total_bill', y='tip', data = tips_df, hue = 'size')
```

Scatter plot showing the relationship between the total_bill and tip variables using sns.relplot(). Seaborn's relplot function is a high-level function used to create relational plots. 

It is a flexible function that can create scatter plots, line plots, or other types of plots to visualize the relationship between variables in your dataset. 

The main advantage of relplot is that it allows you to easily create different types of relational plots by specifying the kind parameter.

```python
sns.relplot(x="time", y="total_bill", kind="line", data=tips_df)
```

Here, we specify kind="line" to create a line plot showing the relationship between time and total_bill variables.

```python
sns.relplot(x = 'total_bill', y='tip', data = tips_df, style ='smoker', hue = 'time')
```

```python
# Create a scatter plot with marker size based on party size
sns.relplot(x="total_bill", y="tip", size="size", data=tips_df)
```

Here, We specify size="size" to adjust the size of markers based on the size variable in the dataset, which represents the size of the dining party.

```python
# Specify a custom range of marker sizes from 20 to 200. 
sns.relplot(x="total_bill", y="tip", size="size", sizes=(20, 200), data=tips_df)
```

## Plot shows the distribution of total_bill amounts for each day of the week using sns.catplot()

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips_df)
```

Seaborn's catplot function is a high-level function used to create categorical plots. It allows you to visualize the *relationship between one categorical variable and one numerical variable or two categorical variables*. catplot is a generalization of several axes-level functions in Seaborn, such as barplot, boxplot, violinplot, and stripplot, among others.

```python
# swarm plot showing the distribution of total_bill amounts for each day of the week.
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="strip", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="point", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="bar", data=tips_df)
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=tips_df)
```

Depending on your data and the relationships you want to visualize, you can choose the appropriate kind of plot to create.

## Violin plot showing the distribution of total bill amounts by day of the week.

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x='day', y='total_bill', data=tips_df, palette='muted')
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

## Bar plot showing the distribution of total bill amounts by day of the week.

It shows the mean or another statistical measure of the variable within each category, along with confidence intervals to indicate the uncertainty in the estimate.

## JointPlot
Combines multiple plots to visualize the relationship between two variables, including scatter plots, histograms, and regression lines.

```python
# Create a joint plot of total bill and tip amount
sns.jointplot(x="total_bill", y="tip", data=tips_df)
```

Seaborn's jointplot function is used to create a multi-panel figure that shows the joint distribution of two variables along with the marginal distributions of each variable separately. It provides a convenient way to visualize the relationship between two variables and the distribution of each variable individually.

```python
sns.jointplot(x = tips.total_bill, y=tips.tip, data = tips, kind="reg")
```

```python
sns.jointplot(x = tips.total_bill, y=tips.tip, data = tips, kind="hex")
```

```python
sns.jointplot(x = tips.total_bill, y=tips.tip, data = tips, kind="kde")
```

## PairPlot

Seaborn's pairplot function is a convenient way to create a grid of pairwise plots for a given DataFrame. It plots pairwise relationships in the dataset, creating scatterplots for numerical variables and histograms for each variable along the diagonal. This allows you to quickly visualize the relationships between multiple variables in the dataset.

```python
sns.pairplot(tips_df)
```

```python
sns.pairplot(tips_df, hue = 'sex')
```

```python
sns.pairplot(tips_df, hue = 'smoker')
```

```python
sns.pairplot(tips_df, hue = 'smoker', palette='coolwarm')
```

```python
sns.pairplot(tips_df, hue = 'smoker', palette='husl')
```

## Heatmap
Visualizes the correlation or relationship between multiple variables using colors.

Seaborn's heatmap function is used to create a heatmap, which is a graphical representation of data where the individual values contained in a matrix are represented as colors. Heatmaps are particularly useful for visualizing the magnitude of relationships or correlations between variables in a dataset.

```python
corr_matrix = tips_df.corr()
```

```python
sns.heatmap(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```

## Countplot

Seaborn's countplot function is used to create a bar plot showing the counts of observations in each category of a categorical variable. It is particularly useful for visualizing the distribution of categorical data and comparing the frequencies of different categories.

```python
sns.countplot(x = tips_df.sex)
```

```python
sns.countplot(x="day", data=tips_df)
```

```python
sns.countplot(y="day", data=tips_df, palette="husl")
```

## LM plot
Seaborn's lmplot (linear model plot) function is used to create scatter plots with regression lines fit to the data. It's particularly useful for visualizing the relationship between two numerical variables and assessing the strength and direction of the relationship.

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips_df)
```

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips_df, aspect=4)
```

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips_df, aspect=2)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="day", data=tips_df, col="day")
```

lmplot is a versatile function that provides a convenient way to create scatter plots with regression lines in Seaborn.
