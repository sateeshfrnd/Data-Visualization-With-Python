# Matplotlib
Matplotlib is a versatile library for creating a wide range of visualizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line Plot
Displaying data points connected by straight line segments.

```python
# Generates an array of evenly spaced values  from 0 to 55Ï€ (inclusive) with a step size of 0.1
x = np.arange(0, 5*np.pi, 0.1)
print(x)
```

```python
plt.plot(x)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### Plot with two subplots arranged horizontally

```python
plt.figure()
# Creates the first subplot in a grid with 1 row and 2 columns, and sets the current subplot to the first one.
plt.subplot(1,2,1) 
plt.plot(x, np.sin(x))

# Creates the second subplot in the same grid, and sets the current subplot to the second one.
plt.subplot(1,2,2) 
plt.plot(x, np.cos(x))

# Displays the entire figure with both subplots.
plt.show()
```

### Plot with two subplots arranged vertically

```python
plt.figure()

# Creates the first subplot in a grid with 2 row and 1 columns, and sets the current subplot to the first one.
plt.subplot(2,1,1)
plt.plot(x, np.sin(x), color='blue', linestyle='-', linewidth=2, label='sin(x)')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()

# Creates the second subplot in the same grid, and sets the current subplot to the second one.
plt.subplot(2,1,2) 
plt.plot(x, np.cos(x), color='red', linestyle='--', linewidth=2, label='cos(x)')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()

# Adjust subplots to prevent overlap
plt.tight_layout() 

plt.show()
```

### Plot with four subplots

```python
plt.figure()
plt.subplot(2,2,1) # one page divide into 4 part and access one part to plot
plt.plot(x, np.sin(x), color='blue', linestyle='-', linewidth=2, marker='o', markersize=5, label='sin(x)')
plt.title('Sine Function', fontsize=10)
plt.xlabel('x', fontsize=8)
plt.ylabel('sin(x)', fontsize=8)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(loc='upper right')

plt.subplot(2,2,2) # one page divide into 4 part and access second part to plot
plt.plot(x, np.cos(x), color='red', linestyle='--', linewidth=2, marker='x', markersize=8, label='cos(x)')
plt.title('Cosine Function', fontsize=10)
plt.xlabel('x', fontsize=8)
plt.ylabel('cos(x)', fontsize=8)


plt.subplot(2,2,3) # one page divide into 4 part and access second part to plot
plt.plot(x, np.tan(x))
plt.title('Tan Function', fontsize=10)
plt.xlabel('x', fontsize=8)
plt.ylabel('tan(x)', fontsize=8)

plt.subplot(2,2,4) # one page divide into 4 part and access second part to plot
plt.plot(x, np.cos(x+1), color='red', linestyle='--', linewidth=2, marker='.', markersize=8, label='cos(x)')
plt.title('Cosine Function(x+1)', fontsize=10)
plt.xlabel('x+1', fontsize=8)
plt.ylabel('cos(x+1)', fontsize=8)
plt.grid(True, linestyle=':', linewidth=0.5, alpha=0.5)
plt.legend(loc='upper right')


# Adjust subplots to prevent overlap
plt.tight_layout() 

plt.show()
```

### Scatter plot with sample data related to customers and their credit scores.

```python
# Create DataFrame
df = pd.DataFrame(
    {
        'Customer': ['Customer1', 'Customer2', 'Customer3', 'Customer4', 'Customer5'],
        'Age': [25, 30, 35, 40, 45],
        'Credit_Score': ['Good', 'Fair', 'Good', 'Poor', 'Fair']
    }
)
df
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(df['Age'], df['Credit_Score'], c=df['Credit_Score'].map(colors), s=100, alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Credit Score')
plt.title('Customer Credit Score by Age')

# Define colors for different credit score categories
colors = {'Good': 'green', 'Fair': 'orange', 'Poor': 'red'}

# Add legend
for credit_score, color in colors.items():
    plt.scatter([], [], c=color, label=credit_score)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Define colors for different credit score categories
colors = {'Good': 'green', 'Fair': 'orange', 'Poor': 'red'}

# Create scatter plot
plt.figure(figsize=(10, 6))  # Set figure size
for score, color in colors.items():
    plt.scatter(df[df['Credit_Score'] == score]['Age'], df[df['Credit_Score'] == score]['Credit_Score'], 
                color=color, label=score, s=100, alpha=0.7, edgecolors='k', marker='o')

# Additional attributes
plt.xlabel('Age')  # X-axis label
plt.ylabel('Credit Score')  # Y-axis label
plt.title('Customer Credit Score by Age')  # Plot title
plt.legend()  # Add legend
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)  # Show grid lines
plt.xlim(20, 50)  # Set x-axis limits
plt.ylim(0, len(colors))  # Set y-axis limits
plt.gca().set_aspect('auto')  # Adjust aspect ratio

plt.show()
```

## Bar Plot
Bar plots are useful for visualizing categorical data, where each bar represents a category and the height of the bar corresponds to the value of that category. 

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 25]  # Corresponding values for each category
```

```python
plt.figure(figsize=(6,3))
plt.bar(categories,values)
```

```python
plt.figure(figsize=(6,3))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(6,6))
bars = plt.bar(categories, values, color='skyblue', width=0.5, align='center', edgecolor='black', linewidth=1.5, alpha=0.8)
# Add labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height}', ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Additional Attributes')
plt.legend(['Values'])
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.show()
```

### Plot a horizontal Bar plot 

```python
plt.figure(figsize=(6,6))
plt.barh(categories, values, color='skyblue', height=0.5, align='center', edgecolor='black', linewidth=1.5, alpha=0.8)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.legend(['Values'])
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.show()
```

## Histogram
Showing the frequency distribution of a continuous variable.

```python
# Generate 100 random ages between 18 and 70
np.random.seed(0)  # For reproducibility
ages = np.random.randint(18, 70, size=100)
```

```python
plt.hist(ages) # default create 10 beans
plt.show()
```

### Change the number of bins

```python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(ages, bins=20, histtype = 'step') 
plt.show()
```

```python
plt.hist(ages, bins=20, histtype = 'barstacked', color = 'green') 
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(ages, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ages')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(ages, bins=10, histtype='stepfilled', density=True, cumulative=False, color='skyblue', edgecolor='black', alpha=0.7)
# Add labels and title
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('Distribution of Customer Ages')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(ages, bins=10, cumulative=True, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Cumulative Count')
plt.title('Cumulative Distribution of Customer Ages')
plt.grid(True)
plt.show()
```

Each bin's height represents the cumulative count of data points up to that bin. The histogram shows how the number of data points accumulates as the age increases, providing insight into the overall distribution of the data.

```python
# Create a normalized histogram
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(ages, bins=10, density=True, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('Normalized Distribution of Customer Ages')
plt.grid(True) # Add grid lines
plt.show()
```

Here 'density=True' ensures that the histogram is normalized such that the total area under the curve sums to 1, representing the probability density function of the data. This allows for easier comparison of distributions and interpretation of relative frequencies.

### Plot Multiple Histograms on the same axes to compare distributions

```python
# Sample data (customer ages)
np.random.seed(0)  # For reproducibility
ages1 = np.random.randint(18, 40, size=100)  # Generate 100 random ages between 18 and 40
ages2 = np.random.randint(30, 70, size=100)  # Generate 100 random ages between 30 and 70
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
# Plot multiple histograms on the same axes
plt.hist([ages1, ages2], bins=10, histtype='bar', orientation='vertical', 
         rwidth=0.8, align='mid', alpha=0.5, label=['Group 1', 'Group 2'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ages')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
# Plot multiple histograms on the same axes
plt.hist([ages1, ages2], bins=10, histtype='bar', orientation='horizontal', 
         rwidth=0.8, align='mid', alpha=0.5, label=['Group 1', 'Group 2'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ages')
plt.legend()
plt.show()
```

## Box Plot (Box-and-Whisker Plot)
- A box plot, also known as a box-and-whisker plot.
- Visualizing the distribution of a continuous variable through five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. 
- It is particularly useful for detecting outliers and understanding the spread and skewness of the data.

```python
# Generate sample data with 100 random data points from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=100)  
```

```python
plt.boxplot(data)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(5, 3)) 
plt.boxplot(data, vert=False)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

### Create a box plot with customizations

```python
plt.figure(figsize=(8, 6))  # Set the figure size

# Customize box plot properties
boxprops = dict(color='blue', linewidth=2, linestyle='--', alpha=0.7)
whiskerprops = dict(color='green', linewidth=2, linestyle='-.', alpha=0.7)
capprops = dict(color='red', linewidth=2, linestyle='-', alpha=0.7)
medianprops = dict(color='black', linewidth=2, linestyle='-', alpha=0.7)

# Create the box plot
plt.boxplot([data],
            notch=True,  # Display a notch at the median
            vert=True,  # Plot vertical boxes
            patch_artist=True,  # Fill the boxes with color
            showmeans=True,  # Show the mean value
            showfliers=True,  # Show outlier data points
            boxprops=boxprops,  # Customize box properties
            whiskerprops=whiskerprops,  # Customize whisker properties
            capprops=capprops,  # Customize cap properties
            medianprops=medianprops,  # Customize median properties
            labels=['Values'])  # Set custom labels for the box plots

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Customized Box Plot')

# Show the plot
plt.grid(True)
plt.show()
```

## Violin Plot
Similar to a box plot but showing the probability density of the data.

```python
np.random.seed(0)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=1, scale=1.5, size=100)
```

```python
plt.figure(figsize=(7, 5))  
plt.violinplot([data1, data2], showmeans=True)
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Violin Plot')
plt.show()
```

```python
plt.figure(figsize=(7, 5))  

# Create the violin plot
parts = plt.violinplot([data1, data2], showmeans=True, showextrema=True, showmedians=True)

# Customize violin plot properties
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')  # Set fill color for the violin body
    pc.set_edgecolor('black')  # Set edge color for the violin body
    pc.set_alpha(0.7)  # Set transparency for the violin body

# mean marker 
parts['cmeans'].set_color('red')  # Set color for the mean marker

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Customized Violin Plot')
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.show()
```

## Area Plot

An area plot, also known as a filled line plot, is a type of plot used to represent quantitative data over a continuous interval or time period. It is similar to a line plot, but the area between the line and the x-axis is filled with color to emphasize the magnitude of the values.

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(7, 5))  # Set the figure size
plt.fill_between(x, y1, color='skyblue', alpha=0.5, label='sin(x)')  # Fill area for y1
plt.fill_between(x, y2, color='salmon', alpha=0.5, label='cos(x)')   # Fill area for y2

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Area Plot')
plt.legend()
plt.grid(True, linestyle=':', linewidth=1.5, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(9, 6))  # Set the figure size
plt.fill_between(x, y1, color='skyblue', alpha=0.5, label='sin(x)')  # Fill area for y1
plt.fill_between(x, y2, color='salmon', alpha=0.5, label='cos(x)')   # Fill area for y2

plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Area Plot', fontsize=14)
plt.legend(loc='upper right', shadow=True)
plt.grid(True, linestyle='--', linewidth=1.5, alpha=0.5)

plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)

# Customize tick labels
plt.xticks(np.arange(0, 11, 1))  # Set x-axis tick marks every 1 unit
plt.yticks(np.arange(-1, 1.1, 0.5))  # Set y-axis tick marks every 0.5 units

plt.show()
```
